```python
language = []
no_speakers = []

num_iter = len(soup2.select('td:nth-child(1)'))
```

```python
num_iter
```

```python
for i in range(25):
    language.append(soup2.select('td:nth-child(1)')[i].get_text())
    no_speakers.append(soup2.select('td:nth-child(2)')[i].get_text())
```
